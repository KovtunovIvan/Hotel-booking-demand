```python
#после исследования данных приходим к выводам: 
#-данные не имеют пропусков
#-в данных есть категориальные признаки
#-необходимо решить задачу бинарной классификации

#произведем предобработку данных
train.loc[:, train.dtypes=='object']
```

```python
train = train.drop(columns=['reservation_status_date', 'country', 'assigned_room_type', 'reserved_room_type'])
test = test.drop(columns=['reservation_status_date', 'country', 'assigned_room_type', 'reserved_room_type'])
```

```python
#закодируем категорильаные данныe
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

label_encoder.fit(train['arrival_date_month'])

train['arrival_date_month'] = label_encoder.transform(train['arrival_date_month'])
test['arrival_date_month'] = label_encoder.transform(test['arrival_date_month'])
```

```python
train.head()
```

```python
test.head()
```

```python
catergorial_features_train = train.loc[:, train.dtypes=='object'].columns
catergorial_features_train
```

```python
catergorial_features_test = test.loc[:, train.dtypes=='object'].columns
catergorial_features_test
```

```python
category = ['hotel', 'meal', 'market_segment', 'distribution_channel']

for i in category:
    label_encoder = LabelEncoder()

    label_encoder.fit(train[i])

    train[i] = label_encoder.transform(train[i])
    test[i] = label_encoder.transform(test[i])
```

```python
train.head()
```

```python
#закодируем часть данных с помощью другого типа кадировщика
train = pd.get_dummies(train, columns=['deposit_type', 'customer_type'])
test = pd.get_dummies(test, columns=['deposit_type', 'customer_type'])
```

```python
train.head()
```

```python
train.dtypes
```

```python
test.dtypes
```

```python
#обучим модель
X = train.drop(columns=['is_canceled'])
y = train['is_canceled']
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X,y)

pred = model.predict(X)
```

```python
#проверим результат на метриках аккуратности, точности и полноты
from sklearn.metrics import accuracy_score, precision_score, recall_score

print(accuracy_score(pred, y))
print(precision_score(pred, y))
print(recall_score(pred, y))
```

```python
#построим матрицу ошибок
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(model, X, y)
plt.show()
```

```python
#предскажем ответ для тествовго набора данных
test_pred = model.predict(test)
```

```python
test_pred = pd.DataFrame(test_pred, columns=['is_canceled'])

test_pred = test_pred.reset_index()

test_pred.to_csv('solution.csv', index=False)
```
